## Exemple pour extraire un échantillon du dump de Wikidata


## 1. Exemple pour afficher une entité

```scala
//import json
//wikiFile = sc.textFile('latest-all.json')

val NUMBER_ENTITY = 2;

val wikiFile = sc.textFile("latest-all.json");

val sample = wikiFile.take(NUMBER_ENTITY+1).drop(1).map(x=>x.dropRight(1));

println("[\n"+sample.mkString(",\n")+"\n]"); 
```

## 2. Injection de 1000 entités dans le fichier Wikidata-sample.json

```scala
import java.io._

val t1 = System.nanoTime

val NUMBER_ENTITY = 1000;

val wikiFile = sc.textFile("latest-all.json");

val sample = wikiFile.map(x=>x.dropRight(1)).take(NUMBER_ENTITY+1).drop(1);


val writer = new PrintWriter(new File("Wikidata-sample_scala.json" ))
writer.write("[\n"+sample.mkString(",\n")+"\n]")
writer.close()

val duration = (System.nanoTime - t1) / 1e9d;

println(duration +" seconds"); //secondes
```
