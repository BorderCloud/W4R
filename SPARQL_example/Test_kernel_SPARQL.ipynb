```sparql
%endpoint http://dbpedia.org/sparql
%display table withtypes

SELECT DISTINCT ?property
WHERE {
   ?s ?property ?person .
   ?person rdf:type foaf:Person .
}
LIMIT 10
```

```sparql
%endpoint http://dbpedia.org/sparql
%display diagram

CONSTRUCT {
    ?p1 dbp:successor ?p2 .
    ?p2 dbp:successor ?p3 .
    ?p3 dbp:successor ?p4 .
}
WHERE {
    ?p1 rdf:type foaf:Person .
    ?p1 dbp:successor ?p2 .
    ?p2 dbp:successor ?p3 .
    ?p3 dbp:successor ?p4 .
}
LIMIT 50
```
